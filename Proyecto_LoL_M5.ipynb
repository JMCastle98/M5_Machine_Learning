# Información necesaria para el análisis del desempeño en eSports

Muchos grupos de interés se ven en la necesidad de tener información proveniente de análisis de eSports, por ejemplo, los desarrolladores de videojuegos buscan información que les permita encontrar patrones en la jugabilidad con la finaldiad de realizar cambios en su juego como lo son mapas, reglas, personajes u objetos (Bozorgzadeh, 2017). Ellos también buscan detectar trampas y de esta forma crear sistemas "_anti-cheats_" que permitan un juego más justo.

Si bien mucha de la información recolectada para su análisis es explotada para fines como el aumento de audiencia, momentos destacados o para pláticas de comentaristas durante y después de las partidas. Un ámbito que puede ser de mayor intéres es el relacionado con el rendimiento de los equipos y jugadores. Por ejemplo, análistas, coaches e incluso los propios jugadores presentan particular interés en el desempeño de sus adversarios, por lo que al igual que en los deportes tradicionales, deben conocer a su enemigo y preparar sus estrategias.

En general nos encontramos con cuatro aspectos de posible análisis en los eSports:

*   **Equipos**: La mayoría de profesionales de eSports participan en juegos que presentan enfrentamientos entre qeuipos, donde el entendimiento de los oponentes y su estrategia juegan un papel crítico.
*   **Jugadores**: Las debilidades y fortalezas de cada integrante del equipo permiten trazar un plan que permita bloquear su desempeño en el rol que ejecuta.
*   **Personajes**: Los roles de cada jugador en los distintos juegos vienen dados por el avatar o personaje elegido, entender las habilidades de cada personaje y sus limitaciones tienen un peso en el desempeño como jugador.
*   **Objetos**: Cada personaje puede equiparse con distintos objetos, entender las diferentes composicions y como adecuarlas a la situación en que encuentras trae un abánico de posibiliades para el desarrollo de una partida.

Está claro que cada uno de estos aspectos brinda un cúmulo de información por partida que solo puede ser explotada por análistas expertos. Hoy en día también se habla de páginas de apuestas, eSports de fantasía, ligas colegiales y más importante aún: Canales de televisión enteramente dedicados a eSports. Detrás de todo esto surge una pregunta importante: ¿Dónde conseguimos la información? Para responder está pregunta podemos empezar por conocer algunas fuentes de esta información que hoy en día es tan preciada.







# Fuentes de datos sobre eSports

Antes de que la información se analizada, debe ser recolectada. Algunas de las fuentes más comúnes donde la información sobre eSports puede ser recolectada son:

*   **APIs**: Para los juegos más importantes, los desarrolladores proveen una interfaz de programación de aplicaciones ([API](https://es.wikipedia.org/wiki/Interfaz_de_programaci%C3%B3n_de_aplicaciones)) con la cual se permite la descarga de datos generados a los usuarios registrados.
*   **Proveedores estadísticos y APIs de terceros**: Ya que no todos los desarrolladores de juegos recolectan los datos, terceros proveen APIs donde guardan sus datos obtenidos mediante la búsqueda en registros de los juegos, analizando repositorios públicos, haciendo _scrapping_ en repeticiones de Twitch o Youtube, o bien pidiendo archivos de repetición a los usuarios.
*   **Repositorios de archivos**: Como mencionamos anteriormente muchos juegos guarda repeticiones de las partidas, pero no necesariamente en formatos de video, sino en formatos especiales que guardan cada acción de una partida y pueden reproducirse de nuevo con el motor del juego para analizar distintos momentos o ver como evoluciona la partida desde uno o múltiples ángulos. Estos archivos pueden ser tratados o simplemente guardarse en crudo en repositorios.


# Método KMeans
        1.- Definimos nuestra distancia euclidiana entre puntos y se dan parámetros de arranque
        2.- Especificamos el número de grupos que se quieren identificar
        3.- Se inicializa el algoritmo
        4.- Se asgina un centroide a cada grupo
        5.- Se hace la actualización

Se realiza un método de codo para saber cuantos centroídes son necesarios:

```python
distorcion=[]
K=range(1,10)

for k in K:
    kmeans= KMeans(n_clusters=k)
    kmeans.fit(X)
    distorcion.append(kmeans.inertia_)
    
plt.plot(K,distorcion, 'bx-')
plt.xlabel('Numero de clusters')
plt.ylabel('Distorción')
plt.title('Método elbow')
plt.show()
```

Reforzamos con el método de la silueta:

```python
from sklearn.metrics import silhouette_score

silhouette = {}
K=range(2,10)

for k in K:
    kmeans= KMeans(n_clusters=k)
    kmeans.fit(X)
    silhouette[k] = silhouette_score(X, kmeans.labels_, metric = 'euclidean')
```

```python
plt.bar(range(len(silhouette)), list(silhouette.values()))
plt.xticks(range(len(silhouette)), list(silhouette.keys()))
plt.grid()
plt.title("score del método de la silueta", fontweight = 'bold')
plt.xlabel("Numero de clusters")
```

Construimos el modelo con 2 clusters:

```python
kmeans= KMeans(n_clusters=2)
modelo=kmeans.fit_predict(X)

centroides=kmeans.cluster_centers_.T

plt.scatter(X[modelo==0,0],X[modelo==0,1], c='red')
plt.scatter(X[modelo==1,0],X[modelo==1,1], c='blue')


plt.scatter(*centroides, marker='x', color='black')


plt.title('Método K-Means para 2 clusters desde Sklearn', size = 18)
plt.xlabel('Componente Principal 1',  size = 14)
plt.ylabel('Componente Principal 2',  size = 14)
plt.show()
```

```python
from sklearn.svm import SVC #Support Vector Classifier
SupportVectorMachine = SVC(C=0.1, random_state = 4)
SupportVectorMachine.fit(X_train, y_train) 
y_pred = SupportVectorMachine.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

def calcularAccuracy(TP, TN, FP, FN):
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    accuracy = accuracy * 100
    return accuracy
def calcularSensibilidad(TP, TN, FP, FN):
    sensibilidad = TP / (TP + FN)
    sensibilidad = sensibilidad * 100
    return sensibilidad
def calcularEspecificidad(TP, TN, FP, FN):
    especificidad = TN / (TN + FP)
    especificidad = especificidad * 100
    return especificidad

resultado = confusion_matrix(y_test, y_pred)
print(resultado)
(TN, FP, FN, TP) = resultado.ravel()
print("True positives: "+str(TP))
print("True negatives: "+str(TN))
print("False positives: "+str(FP))
print("False negative: "+str(FN))

acc = calcularAccuracy(TP, TN, FP, FN)
sen = calcularSensibilidad(TP, TN, FP, FN)
spec = calcularEspecificidad(TP, TN, FP, FN)
print("Precision:"+str(acc)+"%")
print("Sensibilidad:"+str(sen)+"%")
print("Especificidad:"+str(spec)+"%")
```

```python
def revisar_match_data(match_ids):     
    
    buffer = [[]]

    for match_id in match_ids:
        url_match_info = f"https://americas.api.riotgames.com/lol/match/v5/matches/{match_id}?api_key={token}"
        response = requests.get(url_match_info) 
        match_data = response.json() 
        if response.status_code == 200: 
            if (match_data['info']['gameMode'] == "CLASSIC") or (match_data['info']['gameMode'] == "ARAM"):
                if match_id == match_ids[0]: 
                    buffer[0] = match_id
                else: buffer.append(match_id)
    return buffer
```

```python
def lol_match_data(match_ids):     
    serie = [[]]*len(match_ids)
    j = 0

    for match_id in match_ids:
        url_match_info = f"https://americas.api.riotgames.com/lol/match/v5/matches/{match_id}?api_key={token}"
        response = requests.get(url_match_info) 
        #response.status_code
        match_data = response.json() # match_data['info']['gameMode'] 

        participantes = match_data['info']['participants']

        # Ciclo para encontrar el id del equipo al que perteneció el jugador
        for i in range(0,len(participantes)): 
            if participantes[i]['summonerName'] == nickname:
                team_id = participantes[i]['teamId']

        
        for i in range(0,len(match_data['info']['teams'])):
            if match_data['info']['teams'][i]['teamId'] == team_id:
                equipo = match_data['info']['teams'][i]
        
        participantes_equipo = []
        # Pongamos a los participantes del equipo en un arreglo para acceder más fácilmente a su información
        for i in range(0,len(participantes)): 
            if participantes[i]['teamId'] == team_id:
                participantes_equipo.append(participantes[i])
        participantes_equipo = pd.DataFrame(participantes_equipo)
        
        df_match_date = match_data['info']['gameStartTimestamp']
        df_match_date = datetime.fromtimestamp(df_match_date/1000,pytz.timezone("Mexico/General"))
        df_match_date = df_match_date.strftime('%Y-%m-%d %H:%M:%S')

        if participantes_equipo['win'][0] == True:
            df_result = 1
        else: df_result = 0

        df_team_kills = participantes_equipo['kills'].sum()
        df_team_deaths = participantes_equipo['deaths'].sum()
        df_team_assists = participantes_equipo['assists'].sum()

        if equipo['objectives']['champion']['first'] == True:
            df_first_blood = 1
        else: df_first_blood = 0

        tiempo_jugado = participantes_equipo['timePlayed'][0]/60
        df_ckpm = round((df_team_kills+df_team_deaths)/tiempo_jugado,4)

        if equipo['objectives']['dragon']['first'] == True:
            df_first_dragon = 1
        else: df_first_dragon = 0

        df_dragons = participantes_equipo['dragonKills'].sum()

        if equipo['objectives']['riftHerald']['first'] == True:
            df_first_herald = 1
        else: df_first_herald = 0

        df_heralds = equipo['objectives']['riftHerald']['kills']

        if equipo['objectives']['baron']['first'] == True:
            df_first_baron = 1
        else: df_first_baron = 0

        df_barons = equipo['objectives']['baron']['kills']

        if equipo['objectives']['tower']['first'] == True:
            df_first_tower = 1
        else: df_first_tower = 0

        df_towers = equipo['objectives']['tower']['kills']
        df_inhibitors = equipo['objectives']['inhibitor']['kills']
        df_damage_to_champions = participantes_equipo['totalDamageDealtToChampions'].sum()
        df_dpm = round(df_damage_to_champions/tiempo_jugado,4)
        df_damage_taken_per_minute = round((participantes_equipo['totalDamageTaken'].sum())/tiempo_jugado,4)
        df_damage_mitigated_per_minute = round((participantes_equipo['damageSelfMitigated'].sum())/tiempo_jugado,4)
        df_wards_placed = participantes_equipo['wardsPlaced'].sum()
        df_wpm = round(df_wards_placed/tiempo_jugado,4)
        df_wards_killed = participantes_equipo['wardsKilled'].sum()
        df_wcpm = round(df_wards_killed/tiempo_jugado,4)
        df_control_ward_bought = participantes_equipo['visionWardsBoughtInGame'].sum()
        df_vision_score = participantes_equipo['visionScore'].sum()
        df_vspm = round(df_vision_score/tiempo_jugado,4)
        df_total_gold = participantes_equipo['goldEarned'].sum()
        df_gold_spent = participantes_equipo['goldSpent'].sum()
        df_minion_kills = participantes_equipo['totalMinionsKilled'].sum()
        df_cspm = round((participantes_equipo['neutralMinionsKilled'].sum()+participantes_equipo['totalMinionsKilled'].sum())/tiempo_jugado,4)

        serie[j] = {
            'match_date': df_match_date,
            'result':df_result,
            'team_kills':df_team_kills,
            'team_deaths':df_team_deaths, 
            'team_assists':df_team_assists,
            'first_blood':df_first_blood,
            'ckpm':df_ckpm,
            'first_dragon':df_first_dragon,
            'dragons':df_dragons,
            'first_herald':df_first_herald,
            'heralds':df_heralds,
            'first_baron':df_first_baron,
            'barons':df_barons,
            'first_tower':df_first_tower,
            'towers':df_towers,
            'inhibitors':df_inhibitors,
            'damage_to_champions':df_damage_to_champions,
            'dpm':df_dpm,
            'damage_taken_per_minute':df_damage_taken_per_minute,
            'damage_mitigated_per_minute':df_damage_mitigated_per_minute,
            'wards_placed':df_wards_placed,
            'wpm':df_wpm,
            'wards_killed':df_wards_killed,
            'wcpm':df_wcpm,
            'control_ward_bought':df_control_ward_bought,
            'vision_score':df_vision_score,
            'vspm':df_vspm,
            'total_gold':df_total_gold,
            'gold_spent':df_gold_spent,
            'minion_kills':df_minion_kills,
            'cspm':df_cspm
        }
        j=j+1
        
    df = pd.DataFrame.from_dict(serie)    
    return df
```
